```python
import os
import datetime
import random

print("--- Intermediate Python Core Exercise ---")
print("This exercise covers Python fundamentals, data structures, functions,")
print("core libraries (os, datetime, random), file handling, and exception handling.")
print("Estimated completion time: 3-4 hours.")
print("------------------------------------------")

DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor', 'Pressure_Sensor', 'Light_Sensor']
LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']
READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}
# --- Task 1: Python Fundamentals, Lists, Tuples, Sets, Dictionaries, Indexing & Slicing ---
print("\n--- Task 1: Data Generation and Basic Structures ---")
raw_sensor_data = []
for i in range(1, 101):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    device_id=f"DEV{i:03d}"
    device_type=random.choice(DEVICE_TYPES)
    location=random.choice(LOCATIONS)
    min_val,max_val=READING_RANGES[device_type]
    value=round(random.uniform(min_val,max_val),2)
    reading=(timestamp,device_id,device_type,location,value)
    raw_sensor_data.append(reading)
print(f"Successfully generated {len(raw_sensor_data)} readings.")
print("Here is a sample of the first 3 readings:")
for reading in raw_sensor_data[:3]:
    print(reading)
# 1.2 From 'raw_sensor_data', create a list named 'temperature_readings' containing only the readings
temperature_reading=[]
for timestamp,device_id,device_type,location,value in raw_sensor_data:
  if device_type=="Temperature_Sensor":
    reading_dict={'timestamp':timestamp,'device_id':device_id,'location':location,'temperature_celsius':value}
    temperature_reading.append(reading_dict)
# 1.3 Create a set named 'unique_locations' containing all unique locations from 'raw_sensor_data'.
unique_locations = set()
for reading in raw_sensor_data:
    loc = reading[3]
    unique_locations.add(loc)
#1.4 Create a dictionary named 'device_stats' where keys are 'device_type' and values are
device_values = {}
for device_type in DEVICE_TYPES:
    device_values[device_type] = []
for timestamp, device_id, device_type, location, value in raw_sensor_data:
    device_values[device_type].append(value)
device_stats = {}
for device_type, values in device_values.items():
    min_val = min(values)
    max_val = max(values)
    device_stats[device_type] = (min_val, max_val)

print("\n--- Task 1.4 Output ---")
for device_type, min_max in device_stats.items():
    print(f"{device_type}: Min = {min_max[0]}, Max = {min_max[1]}")
#1.5.1 List Slicing: Extract elements from the 10th to the 20th (inclusive) from 'raw_sensor_data'.
print("\n--- Task 1.5: Advanced Indexing and Slicing (Lists and Strings) ---")
sample_string = "PythonProgrammingIsFun"
sliced_readings_1 = raw_sensor_data[9:20]
sliced_readings_2 = raw_sensor_data[::5]
# 1.5.2 String Slicing: Extract the substring "Programming" from 'sample_string'.
substring_1 = sample_string[6:17]
substring_2 = sample_string[5::-1]
print("substring_1:", substring_1)
print("substring_2:", substring_2)

# 1.5.3 Mixed Slicing: Access the 'device_type' of the 5th reading in 'raw_sensor_data' using indexing.

fifth_reading_device_type = raw_sensor_data[4][2]
device_type_suffix = fifth_reading_device_type[-6:]

print("\n--- Task 1.5.3 Output ---")
print("5th Reading Device Type:", fifth_reading_device_type)
print("Device Type Suffix:     ", device_type_suffix)
# --- Task 2: Writing and Using Functions in Python ---
print("\n--- Task 2: Function Definitions and Usage ---")
# 2.1 Write a function `get_readings_by_location(readings_list, target_location)
def get_readings_by_location(readings_list, target_location):
    filtered_readings = []

    for timestamp, device_id, device_type, location, value in readings_list:
        if location == target_location:
            # We add the original reading tuple to our filtered list
            reading = (timestamp, device_id, device_type, location, value)
            filtered_readings.append(reading)

    return filtered_readings
# 2.2 Write a function `calculate_average_temperature(temp_data_list)`

def calculate_average_temperature(temp_data_list):
    # Guard clause: Return 0.0 if list is empty
    if not temp_data_list:
        return 0.0

    total_temp = 0.0
    for reading in temp_data_list:
        total_temp += reading['temperature_celsius']

    avg_temp = total_temp / len(temp_data_list)
    return round(avg_temp, 2)


# Let's test the function using your list from Task 1.2!
avg_temp = calculate_average_temperature(temperature_reading)
print(f"Average Temperature: {avg_temp}°C")
```
