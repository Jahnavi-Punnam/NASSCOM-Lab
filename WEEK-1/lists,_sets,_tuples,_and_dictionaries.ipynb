```python
#lists
fruits=['apple','banana','cherry','apple']
print("original list:",fruits)
print("first element:",fruits[0])
```

```python
fruits.append('orange')
print("List after appending:",fruits)
```

```python
fruits.remove('banana')
print("list after removing:",fruits)
print("Number of fruits:",len(fruits))
```

```python
#tuple
dimensions=(10,20,30)
print("Tuple:",dimensions)
```

```python
print("second element:",dimensions[1])
```

```python
#sets
unique_num={1,2,3,2,1}
print("set(no duplicates):",unique_num)
```

```python
unique_num.add(4)
print("set after adding:",unique_num)
```

```python
unique_num.remove(2)
print("set after removing:",unique_num)
```

```python
set_a={1,2,3}
set_b={3,4,5}
print("union:",set_a.union(set_b))
print("intersection:",set_a.intersection(set_b))
print("difference:",set_a.difference(set_b))
```

```python
#dictionaries
student={'name':'Alice','age':21,'course':'python'}
print("dictionary:",student)
```

```python
print("student name:",student['name'])
```

```python
student['grade']='A'
print("dict after adding grade:",student)
del student['age']
print("dict after deleting grade:",student)
for key,value in student.items():
  print(f"{key}:{value}")
```

```python
keys=['id','name','marks']
value=[101,'John',88]
student_inf=dict(zip(keys,value))
print("student info:",student_inf)
```
