```python
def greet():
  print("Hello! Welcome to python functios")
greet()
```

```python
def greet_user(name):
  print(f"Hello,{name}")
greet_user("Jahnavi")
greet_user("Theo")
```

```python
def greet_user(name="Guest"):
  print(f"Welcome,{name}!")
greet_user()
greet_user("Jahnavi")
```

```python
def print_names(*names):
  print("Names passed:",names)
  for name in names:
    print("Hello",name)
print_names('Jahnavi','Alice','Bob')
```

```python
def print_info(**info):
  print("User Info:")
  for key,value in info.items():
    print(f"{key}:{value}")
print_info(Name="Jahnavi",Age=18,City="Hyderabad")
```

```python
def show_details(*args,**kwargs):
  print("Args:",args)
  print("Kwargs:",kwargs)
show_details(1,2,3,Name="Jahnavi",Age=18,City="Hyderabad")
```

```python
def multiply(a,b):
  return a*b
def square(n):
  return multiply(n,n)
print("Square of 4:",square(4))
```
